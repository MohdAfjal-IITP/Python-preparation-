```python
# ✅ Q6 (Intermediate): Find the largest and smallest elements in a tuple.
my_tuple = (10, 20, 60, 10, 5, 100)

def find_min_max (my_tuple):
  smallest = my_tuple[0]
  largest= my_tuple[0]
  for num in my_tuple:
   if num < smallest :
    smallest = num
   if num > largest:
    largest = num
  return smallest,largest
smallest,largest = find_min_max (my_tuple)
print("Smallest element:", smallest)
print("Largest element:", largest)


```

```python
# ✅ Q23 (Basic): Find the union, intersection, and difference of two sets.
Set1 = {"Afjal" , 5 , 0 , 6 , "khan"}
Set2 = {'Ashraf','Afjal',5 , 7 , 'mohd'}
x = Set1.union(Set2)
y = Set1.intersection(Set2)
z = Set1.difference(Set2)
w = Set2.difference(Set1)
print(f'Union set1 to set2 : {x}')
print(f'intersection set1 to set2 : {y}')
print(f'difference set1 to set2 : {z}')
print(f'difference set2 to set1 : {w}')
```

```python
# Define two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Perform set operations
union_set = set1 | set2  # Union (all unique elements)
intersection_set = set1 & set2  # Intersection (common elements)
difference_set = set1 - set2  # Difference (elements in set1 but not in set2)

# Print results
print(f"Union: {union_set}")
print(f"Intersection: {intersection_set}")
print(f"Difference (set1 - set2): {difference_set}")
```

```python
# ✅ Q24 (Basic): Check if a set is a subset of another set.


def is_subset(Set1,Set2):


  return Set2.issubset(Set1)

# Example usage:
Set2 = {1,2,3}
Set1 = {1, 2, 3, 4, 5}
Set3 = {1, 2, 6}


print(f'Is {Set2} a subset of {Set1} ? {is_subset(Set1,Set2)}') #  Output: True
print(f"Is {Set3} a subset of {Set1} ? {is_subset(Set3, Set1)}")  # Output: False
```
