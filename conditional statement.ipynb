```python
def count_fun (n):
  i = 1
  while i<=n:
    yield i  # yield ---> mean produce a data.
    i = i+1
```

```python
def fibo_num (n):

  x = 1
  a,b=0,1
  while x<=n:
    yield a
    x +=1
    a,b = b,a+b
```

```python
fibo_num(5)
```

```python
for i in fibo_num(5):
  print(i)
```

```python
def read_file(file_name):
  with open(file_name) as cod :
    for i in cod:
      yield i
```
