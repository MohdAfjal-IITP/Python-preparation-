```python
# calculater --> +,-,*,/
first = int (input("enter first number --> "))
operator = input("enter operator(+ , - , * , / )")
secound = int (input("enter secounh number --> "))

if operator =="+":
   print(first + secound)
elif operator =="-":
  print(first - secound)
elif operator == "*":
  print(first * secound)
elif operator == "/":
  print(first / secound)

else:
   print ('invalid operation')

```
